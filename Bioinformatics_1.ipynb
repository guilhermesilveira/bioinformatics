```
from itertools import combinations, product, permutations
from collections import Counter

def Mismatches(Text, k, max_d):
  mismatches = []
  for d in range(max_d + 1):
    # print(d)
    alt_bases = {'A':'CGT', 'C':'AGT', 'G':'ACT', 'T':'ACG'}
    p = list(combinations(range(k), d))
    # print(p)
    for starting in range(len(Text) - k + 1):
      partial = Text[starting: starting+k]
      for indexes in p:
        to_combine = ([alt_bases[partial[i]] for i in indexes])
        combs = product(*to_combine)
        new_child = list(partial)
        # print(partial)
        for combination in combs:
          for idx, sub in zip(indexes, combination):
            new_child[idx] = sub
          mismatches.append(''.join(new_child))
  return mismatches

# Write your FrequentWordsWithMismatches() function here, along with any subroutines you need.
# Your function should return a list.
def FrequentWordsWithMismatches(Text, k, max_d):
  mismatches = Mismatches(Text, k, max_d)
  return most_frequent(mismatches)


```

```
def MotifSingleEnumeration(Genome, k, d):
  patterns = set()
  for kmer in kmers(Genome, k):
    for neigh in Neighbors(kmer, d):
      patterns.add(neigh)
  return list(patterns)

def MotifEnumeration(Dna, k, d):
  all_kmers = [set(MotifSingleEnumeration(g, k, d)) for g in Dna]
  return set.intersection(*all_kmers)

```

```
def d_single(Pattern, Sequence):
  min_distance = len(Pattern) + 1
  for i in range(len(Sequence) - len(Pattern) + 1):
    d = HammingDistance(Sequence[i:len(Pattern)+i], Pattern)
    min_distance = min(d, min_distance)
  return min_distance

def DistanceBetweenPatternAndStrings(Pattern, Dna):
  distance = 0
  for g in Dna:
    distance += d_single(Pattern, g)
  return distance

def MedianString(k, Dna):
  min_distance = k * len(Dna)
  best_pattern = None
  for i in range(4 ** k):
    pattern = NumberToPattern(i, k)
    distance = DistanceBetweenPatternAndStrings(pattern, Dna)
    if distance < min_distance:
      min_distance = distance
      best_pattern = pattern
  return best_pattern
```

```
actg_map = {"A" : 0, "C" : 1, "G" : 2, "T" : 3}

def ProfileScore(Text, Profile):
  p = 1
  for i in range(len(Text)):
    nucleotide = Text[i]
    nucleotide_index = actg_map[nucleotide]
    p *= Profile[nucleotide_index][i]
  return p

def ProfileMostProbable(Genome, k, Profile):
  max_score = -1
  best = None
  for i in range(len(Genome) - k + 1):
    text = Genome[i: i + k]
    score = ProfileScore(text, Profile)
    if score > max_score:
      max_score = score
      best = text
  return best
```

```
def ScoreMotifs(motifs):
  score = 0
  for i in range(len(motifs[0])):
    appearances = [motifs[j][i] for j in range(len(motifs))]
    common = Counter(appearances).most_common()
    most_common_frequency = common[0][1]
    # print(most_common_frequency)
    score += len(motifs) - most_common_frequency
  return score
```

```
def Profile(motifs):
  profile = [[],[],[],[]]
  total_motifs = range(len(motifs))
  # print(total_motifs)
  for i in range(len(motifs[0])):
    appearances = [motifs[j][i] for j in total_motifs]
    frequencies = Counter(appearances).most_common(4)
    frequencies = dict(frequencies)
    profile[0].append(frequencies.get('A', 0) / len(motifs))
    profile[1].append(frequencies.get('C', 0) / len(motifs))
    profile[2].append(frequencies.get('G', 0) / len(motifs))
    profile[3].append(frequencies.get('T', 0) / len(motifs))
  return profile
```

```
def GreedyMotifSearch(k, t, Dna):
  best_motifs = [sequence[:k] for sequence in Dna]
  best_score = ScoreMotifs(best_motifs)
  # print(motifs)
  for kmer in kmers(Dna[0], k):
    motifs = [kmer]
    # print(f"iniciando com {motifs}")
    for i in range(1, t):
      strand = Dna[i]
      profile = Profile(motifs)
      # print(f"got {profile}")
      motif = ProfileMostProbable(strand, k, profile)
      motifs.append(motif)
      # print(f"agora com {motifs}")
    new_score = ScoreMotifs(motifs)
    if new_score < best_score:
      # print(f"new best {motifs} versus {best_motifs} score {new_score} x {best_score}")
      best_motifs = motifs.copy()
      best_score = new_score
  return best_motifs
```

```
def Profile(motifs, laplace = False):
  profile = [[],[],[],[]]
  all_motifs = range(len(motifs))
  # print(total_motifs)

  len1 = len(motifs) + laplace

  for i in range(len(motifs[0])):
    appearances = [motifs[j][i] for j in all_motifs]
    frequencies = Counter(appearances).most_common(4)
    frequencies = dict(frequencies)
    profile[0].append((frequencies.get('A', 0) + laplace) / len1)
    profile[1].append((frequencies.get('C', 0) + laplace) / len1)
    profile[2].append((frequencies.get('G', 0) + laplace) / len1)
    profile[3].append((frequencies.get('T', 0) + laplace) / len1)
  return profile
```

```
import itertools

def pick_random_motifs(Dna, k):
  max_position = int(np.random.rand()*(len(Dna[0])-k+1))
  best_motifs = [sequence[max_position:max_position + k] for sequence in Dna]
  best_score = ScoreMotifs(best_motifs)
  return best_motifs, best_score

def GibbsSampler(k, t, Dna, N=2000):
  best_motifs, best_score = pick_random_motifs(Dna, k)
  for j in range(N):
    previous_motifs = best_motifs.copy()

    i = np.random.randint(t)
    popped = previous_motifs.pop(i)
    profile = Profile(previous_motifs, laplace=True)
    # print(i,best_motifs, popped,profile, len(previous_motifs), previous_motifs)
    motifi = ProfileMostProbable(Dna[i], k, profile)
    # print("motifi", popped, motifi)
    motifs = list(itertools.chain(best_motifs[:i], [motifi], best_motifs[i+1:]))
    # print(motifs)
    new_score = ScoreMotifs(motifs)
    if new_score <= best_score:
      # print(f"new best {motifs} versus {best_motifs} score {new_score} x {best_score}")
      best_motifs = motifs.copy()
      best_score = new_score
    else:
      return best_motifs.copy(), best_score
  return best_motifs.copy(), best_score
```
